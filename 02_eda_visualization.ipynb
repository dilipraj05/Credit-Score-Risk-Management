```python
# 02_eda_visualization.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the cleaned dataset
df = pd.read_csv("cleaned_credit_data.csv")
```

```python
# Set style
sns.set(style="whitegrid")
```

```python
# How many people defaulted vs. not?
sns.countplot(x='SeriousDlqin2yrs', data=df)
plt.title("Loan Default Distribution (0 = No Default, 1 = Default)")
plt.show()

# Show value counts
print(df["SeriousDlqin2yrs"].value_counts(normalize=True))
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Age Distribution of Applicants")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MonthlyIncome"], kde=True, bins=50)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="SeriousDlqin2yrs", y="DebtRatio", data=df)
plt.title("Debt Ratio vs Default Status")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
